```python
# Analysing the 'Age' feature
def plotAge(data, axes):
    """
    Creates two subplots for age analysis: a KDE plot and a disease probability bar plot.

    Parameters:
    data (DataFrame): The pandas DataFrame containing the data.
    axes (list): A list of two matplotlib axes objects for the subplots.
    """
    # KDE plot
    sns.kdeplot(
        data=data,
        x="age",
        hue="target",
        fill=True,
        ax=axes[0]
    )
    axes[0].set_xlabel('Age')
    axes[0].set_ylabel('Density')
    axes[0].set_title('Age Distribution by Heart Disease Status')
    axes[0].legend(title='Heart Disease', labels=['No Disease', 'Disease'])

    # Bar plot of disease probability by age
    avg = data.groupby("age", as_index=False)["target"].mean()
    sns.barplot(
        data=avg,
        x="age",
        y="target",
        ax=axes[1],
        color='skyblue'
    )
    axes[1].set_xlabel('Age')
    axes[1].set_ylabel('Disease Probability')
    axes[1].set_ylim(0, 1)
    axes[1].set_title('Probability of Heart Disease by Age')
    axes[1].tick_params(axis='x', rotation=45)
```

```python
# Create the plot
fig_age, axes = plt.subplots(nrows=2, ncols=1, figsize=(15, 10))
plotAge(dataset, axes)  # Fixed: using 'dataset' instead of 'data'
plt.tight_layout()
plt.show()
```

```python
# Define categorical and continuous variables
categorical_features = [('sex', ['female', 'male']),
              ('cp', ['typical angina', 'atypical angina', 'non-anginal pain', 'asymptomatic']),
              ('fbs', ['fbs > 120mg', 'fbs < 120mg']),
              ('restecg', ['normal', 'ST-T wave', 'left ventricular']),
              ('exang', ['yes', 'no']),
              ('slope', ['upsloping', 'flat', 'downsloping']),
              ('thal', ['normal', 'fixed defect', 'reversible defect'])]

continuous_features = [('trestbps', 'blood pressure in mm Hg'),
              ('chol', 'serum cholestoral in mg/d'),
              ('thalach', 'maximum heart rate achieved'),
              ('oldpeak', 'ST depression by exercise relative to rest'),
              ('ca', '# major vessels: (0-3) colored by flourosopy')]

# Create subplots for categorical features
fig_categorical, axes = plt.subplots(nrows=len(categorical_features), ncols=3, figsize=(20, 35))

for i, (attribute, labels) in enumerate(categorical_features):
    # Count plot
    sns.countplot(x=attribute, data=dataset, ax=axes[i][0])
    axes[i][0].set_title(f'Distribution of {attribute}')

    # Count plot by target
    sns.countplot(x='target', hue=attribute, data=dataset, ax=axes[i][1])
    axes[i][1].set_title(f'{attribute} Distribution by Target')
    axes[i][1].legend(title=attribute, labels=labels)

    # Mean target value by category
    avg = dataset[[attribute, 'target']].groupby([attribute], as_index=False).mean()
    sns.barplot(x=attribute, y='target', data=avg, ax=axes[i][2])
    axes[i][2].set_title(f'Heart Disease Probability by {attribute}')
    axes[i][2].set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

```python
# Create subplots for continuous features
fig_continuous, axes = plt.subplots(nrows=len(continuous_features), ncols=2, figsize=(15, 25))

for i, (attribute, xlabel) in enumerate(continuous_features):
    # Distribution plot
    sns.histplot(dataset[attribute], ax=axes[i][0], kde=True)
    axes[i][0].set_xlabel(xlabel)
    axes[i][0].set_ylabel('Density')
    axes[i][0].set_title(f'Distribution of {attribute}')

    # Violin plot by target
    sns.violinplot(x='target', y=attribute, data=dataset, ax=axes[i][1])
    axes[i][1].set_xlabel('Target (0 = No Disease, 1 = Disease)')
    axes[i][1].set_ylabel(xlabel)
    axes[i][1].set_title(f'{attribute} by Heart Disease Status')

plt.tight_layout()
plt.show()
```

```python
# Heart Disease Frequency for ages
plt.figure(figsize=(20, 6))
pd.crosstab(dataset.age, dataset.target).plot(kind="bar", figsize=(20,6))
plt.title('Heart Disease Frequency for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(['No Disease', 'Disease'])
plt.show()

# Heart Disease frequency for sex
plt.figure(figsize=(10, 6))
pd.crosstab(dataset.sex, dataset.target).plot(kind="bar", color=['blue','red'])
plt.title('Heart Disease Frequency for Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.xticks(rotation=0)
plt.legend(["No Disease", "Disease"])
plt.ylabel('Frequency')
plt.show()

# Correlation plot for numerical variables
numerical_vars = ['age','trestbps','chol','thalach','oldpeak','ca']

plt.figure(figsize=(10, 8))
df_corr = dataset[numerical_vars]
corr = df_corr.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=.1, square=True)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
# IV. Train Test split
from sklearn.model_selection import train_test_split

# Prepare features and target
predictors = dataset.drop("target", axis=1)
target = dataset["target"]

# Split the data (70% training, 30% testing)
X_train, X_test, Y_train, Y_test = train_test_split(predictors, target, test_size=0.30, random_state=42, stratify=target)

print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
print(f"Training target distribution:\n{pd.Series(Y_train).value_counts()}")
print(f"Testing target distribution:\n{pd.Series(Y_test).value_counts()}")
```
